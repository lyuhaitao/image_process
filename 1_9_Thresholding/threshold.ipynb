```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
args = {'img1':r'../data/DJI_0027.png'}
image = cv2.imread(args['img1'])
image.shape
```

```python
args['img2'] = r'../data/DJI_0027.JPG'
im0027 = cv2.imread(args['img2'])
im0027.shape
```

```python
# the pixel valuess less than T are set to the output value
# use only one channel
(T, threshInv) = cv2.threshold(image[:,:,0], 200, 255, cv2.THRESH_BINARY_INV)
fig = plt.figure(figsize=(10,8))
plt.imshow(threshInv, cmap='gray')
plt.show()
```

```python
# the pixel valuess less than T are set to the output value
(T, threshInv) = cv2.threshold(image, 200, 255, cv2.THRESH_BINARY_INV)
fig = plt.figure(figsize=(10,8))
plt.imshow(threshInv)
plt.show()
```

```python
(T, thresh) = cv2.threshold(image[:,:,0], 200, 255, cv2.THRESH_BINARY)
fig = plt.figure(figsize=(10,8))
plt.imshow(thresh)
plt.show()
```

```python
# Masking - gray image
masked = cv2.bitwise_and(image, image, mask=threshInv)
fig = plt.figure(figsize=(10,8))
plt.imshow(masked)
plt.show()
```

```python
# Otsu's thresholding
blurred = cv2.GaussianBlur(image[:,:,0], (7,7), 0)
print('The shape: {}'.format(blurred.shape))
fig = plt.figure(figsize=(10,8))
plt.imshow(blurred, cmap='gray')
plt.show()
```

```python
gray = cv2.cvtColor(im0027,cv2.COLOR_BGR2GRAY)
fig = plt.figure(figsize=(10,8))
plt.imshow(gray,cmap='gray')
plt.show()
```

```python
blurred = cv2.GaussianBlur(gray, (7,7), 0)
fig = plt.figure(figsize=(10,8))
plt.imshow(blurred, cmap='gray')
plt.show()
```

```python
(T, threshInv) = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY_INV | cv2.THRESH_OTSU)
fig = plt.figure(figsize=(10,8))
plt.imshow(threshInv, cmap='gray')
plt.show()
print("INFO: otsu's thresholding value: {}".format(T))
```

```python
(T, threshInv) = cv2.threshold(blurred, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
fig = plt.figure(figsize=(10,8))
plt.imshow(threshInv, cmap='gray')
plt.show()
print("INFO: otsu's thresholding value: {}".format(T))
```
