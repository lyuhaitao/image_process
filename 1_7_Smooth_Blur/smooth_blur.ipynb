* Simple average blurring  
* Gaussian blurring  

```python
import cv2
import matplotlib.pyplot as plt
```

```python
args = {'img1':r'../data/DJI_0027.png'}
image = cv2.imread(args['img1'])
kernelSizes = [(3,3), (9,9), (15,15)]
image.shape
```

```python
fig, axes = plt.subplots(3,1, figsize=(10,24))
for i, ks in enumerate(kernelSizes):
    kx, ky = ks
    blurred = cv2.blur(image, (kx,ky))
    axes[i].imshow(blurred)
    axes[i].set_title('Average: ({}, {})'.format(kx, ky))
plt.show()
```

```python
# Gaussian Bluring
fig, axes = plt.subplots(3,1, figsize=(10,24))
for i, ks in enumerate(kernelSizes):
    kx, ky = ks
    blurred = cv2.GaussianBlur(image, (kx,ky),0)
    axes[i].imshow(blurred)
    axes[i].set_title('Gaussian: ({}, {})'.format(kx, ky))
plt.show()
```

```python
# Median Bluring
fig, axes = plt.subplots(3,1, figsize=(10,24))
for i, ks in enumerate([3,9,15]):
    blurred = cv2.medianBlur(image, ks)
    axes[i].imshow(blurred)
    axes[i].set_title('Median: {}'.format(ks))
plt.show()
```

```python
# Bilateral Bluring
fig, axes = plt.subplots(4,1, figsize=(10,32))
axes[0].imshow(image)
axes[0].set_title('original')
params = [(11,21,7), (11,41,21), (11,61,39)]
for i, ps in enumerate(params):
    diameter, sigmaColor, sigmaSpace = ps
    blurred = cv2.bilateralFilter(image, diameter, sigmaColor, sigmaSpace)
    axes[i+1].imshow(blurred)
    axes[i+1].set_title('Blurred d={}, sc={}, ss={}'.format(diameter,sigmaColor, sigmaSpace))
plt.show()
```
